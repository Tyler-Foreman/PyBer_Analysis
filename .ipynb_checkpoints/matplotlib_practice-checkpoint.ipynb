```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, label='Fare ($)')
plt.xlabel('Month')
plt.ylabel('USD ($)')
plt.ylim(0,75)
plt.legend()
plt.grid()
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="D", color='g', label='Boston')
ax.legend()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label="Boston")
ax.set_title('PyBer Fare by Month')
#Add a legend for the city of Boston, NEED HELP with this one.
ax.legend()
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="s", color="green", lw=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
plt.show()
```

```python
plt.scatter(y_axis, x_axis, c='r', label='Chicago')
plt.ylabel('Month')
plt.gca().invert_yaxis()
plt.xlabel('Fare ($)')
plt.xlim(0,50,10)
plt.title('PyBer Fare by Month')
plt.legend()
plt.show()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='Skyblue', alpha=0.8, edgecolor='black', linewidths=2, label='Boston')
ax.legend()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, colors=colors, explode=explode_values, autopct='%1.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
ax.pie(y_axis,labels=x_axis, colors=colors, autopct='%1.1f%%', explode=explode_values, shadow=True, startangle=95, counterclock=False)
plt.show()
```
